# Limits of Diversification
*The anatomy of a Cap weighted portfolio*

```python
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import edhec_risk_kit as erk
import matplotlib.pyplot as plt

np.set_printoptions(suppress =  True)
ind_returns = erk.get_ind_returns()
ind_size = erk.get_ind_size()
ind_nfirms = erk.get_ind_nfirms()
```

```python
print(ind_returns.shape)
print(ind_nfirms.shape)
print(ind_size.shape)
```

```python
ind_mktcap = ind_nfirms * ind_size
```

```python
total_mktcap = ind_mktcap.sum(axis = 'columns')
total_mktcap.plot(figsize = (12,6))
```

```python
cap_weights = ind_mktcap.div(total_mktcap, axis = 'rows')
cap_weights.head()
```

```python
total_market_return  = (cap_weights * ind_returns).sum(axis = 'columns')
```

```python
total_market_index = erk.drawdown(total_market_return).Wealth
total_market_index.plot(figsize = (12,6), title = 'Total Market Wealth Index 1926 - 2018')
```

```python
total_market_index['1980':].plot(figsize = (12,6))
total_market_index['1980':].rolling(window = 36).mean().plot()

```

```python
tmi_tr36rets = total_market_return.rolling(window = 36).aggregate(erk.annualize_rets,12)
tmi_tr36rets.plot(figsize = (12,6), label = 'Tr 36 Month Return', legend = True)
#total_market_return.plot(label= 'return', legend = True)
```

### Rolling Correlations along Multiindex and `.group_by`

```python
ts_corr = ind_returns.rolling(window = 36).corr()
ts_corr.tail()
```

```python
ts_corr.index.names = ['date','industry']
ts_corr.tail()
```

```python
ind_36_corr = ts_corr.groupby(level = 'date').apply(lambda cormat:cormat.values.mean())
```

```python
ind_36_corr.plot()
```

```python
tmi_tr36rets['2007':].plot(label = 'Trailing 36 month return', figsize = (12,6), legend = True)
ind_36_corr['2007':].plot(label = 'Trailing 36 month Correlations', legend = True, secondary_y = True)
```
